```python
# Reload the dataset (if necessary)
df = pd.read_csv(r"C:\Users\HP\Desktop\UPI Fraud Detection\PS_20174392719_1491204439457_log.csv")

# Drop columns with unique values without reassigning
drop = ['amount', 'nameOrig', 'oldbalanceOrg', 'newbalanceOrig', 'nameDest', 'oldbalanceDest', 'newbalanceDest']
df.drop(columns=drop, inplace=True)

# Display the remaining columns
print(df.columns)
```

```python
df.head()
```

```python
 #Data Cleaning
    
# Group by 'type' and sum the 'isFraud' column to get the count of fraudulent transactions by type
fraud = df.groupby('type')['isFraud'].sum().sort_values(ascending=False)

# Display the result
print(fraud)
```

```python
# Reformat the grouped data into a DataFrame for displaying similar to df.head()
fraud_df = pd.DataFrame(fraud).reset_index()

# Rename the columns for better readability
fraud_df.columns = ['Transaction Type', 'Fraudulent Transactions']

# Display the result in a table format
fraud_df.head()
```

```python
# Create a DataFrame for fraudulent transactions by type
fraud_df = pd.DataFrame(fraud).reset_index()
fraud_df.columns = ['type', 'Fraudulent Transactions']

# Merge this DataFrame with the original DataFrame (df)
df_merged = df.merge(fraud_df, on='type', how='left')

# Display the first few rows of the merged DataFrame
df_merged.head()
```

```python
fraud_by_type = df.groupby('type')['isFraud'].sum().reset_index()

# Plot using seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='type', y='isFraud', data=fraud_by_type, palette='Blues_d')

# Add labels and title
plt.title('Number of Fraudulent Transactions by Transaction Type', fontsize=14)
plt.xlabel('Transaction Type', fontsize=12)
plt.ylabel('Number of Fraudulent Transactions', fontsize=12)

# Display the plot
plt.show()
```

```python
df.head()
```

```python
fraud_percentage = df.groupby('type')['isFraud'].mean() * 100
plt.figure(figsize=(10,6))
sns.barplot(x=fraud_percentage.index, y=fraud_percentage.values, palette='coolwarm')
plt.title('Percentage of Fraudulent Transactions by Transaction Type', fontsize=14)
plt.xlabel('Transaction Type', fontsize=12)
plt.ylabel('Percentage of Fraudulent Transactions (%)', fontsize=12)
plt.show()
```

```python
fraud_by_step = df.groupby('step')['isFraud'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(x='step', y='isFraud', data=fraud_by_step, color='blue')
plt.title('Fraudulent Transactions Over Time (Step)', fontsize=14)
plt.xlabel('Step (Time)', fontsize=12)
plt.ylabel('Number of Fraudulent Transactions', fontsize=12)
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['type'], drop_first=True)
```

```python
fraud_rate = df.groupby('type')['isFraud'].mean()
df['fraud_rate_by_type'] = df['type'].map(fraud_rate)
```

```python
df_encoded = pd.get_dummies(df, columns=['type'], drop_first=True)

# Train-test split
X = df_encoded.drop(columns=['isFraud', 'isFlaggedFraud'])
y = df_encoded['isFraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
fraud_percentage = df.groupby('type')['isFraud'].mean() * 100
plt.figure(figsize=(10,6))
sns.barplot(x=fraud_percentage.index, y=fraud_percentage.values, palette='coolwarm')
plt.title('Percentage of Fraudulent Transactions by Transaction Type', fontsize=14)
plt.xlabel('Transaction Type', fontsize=12)
plt.ylabel('Percentage of Fraudulent Transactions (%)', fontsize=12)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
